```python
import pandas as pd
import ast
import os
import glob
import re
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
input_csv_filename = "./input_csv/input_csv.csv"
```

```python
df = pd.read_csv(input_csv_filename)
```

```python
# ----- PLOT 1: F1 Macro per Dataset -----
plt.figure(figsize=(10, 6))
sns.boxplot(x='dataset', y='f1_macro', data=df)
plt.title('F1 Macro per Dataset')
plt.ylabel('F1 Macro')
plt.xlabel('Dataset')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ----- PLOT 2: Average accuracy per Dataset -----
plt.figure(figsize=(10, 6))
mean_acc = df.groupby('dataset')['accuracy'].mean().sort_values()
sns.barplot(x=mean_acc.index, y=mean_acc.values)
plt.title('Average accuracy per Dataset')
plt.ylabel('Accuracy')
plt.xlabel('Dataset')
plt.ylim(0.5, 1.05)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ----- PLOT 3: F1 Macro per Task -----
plt.figure(figsize=(12, 6))
sns.lineplot(x='task', y='f1_macro', hue='dataset', marker='o', data=df)
plt.title('F1 Macro per Task')
plt.ylabel('F1 Macro')
plt.xlabel('Task')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ----- PLOT 4: Precision, Recall and F1 Weighted per Dataset (Boxplot) -----
for metric in ['precision', 'recall', 'f1_weighted']:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='dataset', y=metric, data=df)
    plt.title(f'{metric.capitalize()} per Dataset')
    plt.ylabel(metric.capitalize())
    plt.xlabel('Dataset')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
